```python
y = df['Recommended']
X = df['Review']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
label_mapping = {'Negative': 0, 'Positive': 1}
y_train_binary = [label_mapping[label] for label in y_train]
y_test_binary = [label_mapping[label] for label in y_test]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(X_train)
```

```python
X_train_tfidf = tfidf.transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
X_train_tfidf
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train_tfidf,y_train_binary)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=1000)
log.fit(X_train_tfidf,y_train_binary)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
svc.fit(X_train_tfidf,y_train_binary)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report,accuracy_score
```

```python
def report(model):
    preds = model.predict(X_test_tfidf)
    print(classification_report(y_test_binary,preds))
    print(accuracy_score(y_test_binary,preds))
    plot_confusion_matrix(model,X_test_tfidf,y_test_binary)
    fpr, tpr, thresholds = roc_curve(y_test_binary, preds)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
    
```

```python
print("NB MODEL")
report(nb)
```

```python
print("Logistic Regression")
report(log)
```

```python
print('SVC')
report(svc)
```
